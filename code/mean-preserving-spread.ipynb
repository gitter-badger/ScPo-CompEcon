# mean preserving spreads

We look at mean-preserving spreads in this notebook. It's very useful to plot the pdf of 2 associated distrbutions to get a better understanding of this.

Suppose we have two distributions with the same mean. They differ along an index $r$ which alters the spread of the distribution, i.e. we have 

$$ \int_0^B (F(\theta,r_1) - F(\theta,r_2)) d\theta = 0 $$

where $B$ is a finite upper bound, and we assume $\theta\geq0$

```julia
using Distributions
using Gadfly
# two distributions with the same mean
f1 = Normal(3,0.5)
f2 = Normal(3,1.5)
pdf1(x) = pdf(f1,x)
pdf2(x) = pdf(f2,x)
cdf1(x) = cdf(f1,x)
cdf2(x) = cdf(f2,x)
pdfs = plot([pdf1,pdf2],-1,7,Theme(line_width=1mm),Guide.title("Pdfs"))
```

```julia
cdfs = plot([cdf1,cdf2],-1,7,Theme(line_width=1mm),Guide.title("Cdfs"))
```
